```python
#In these example we are seeing that --> whats the error rate of the algorithms which are used to train the data 
#here we are taking 2 algorithms randomforestregressor() and linearregression()
import pandas as pd
df = pd.read_csv('/content/winequality-red.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,0:-1]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#block 1
#here we are taking randomforestregressor algorithm. 
#in that we finding out the mean square error and mean absolute error then we are comparing these error rates with the error rates of the other algorithm 
#which on of the error rate is less we conclude that algo.
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
```

```python
rfr.fit(x_train,y_train)
```

```python
y_pred=rfr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mae=mean_absolute_error(y_test,y_pred)
```

```python
mae
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
#block 2
#here we are taking linearregression algorithm
#same taking out the error rates from these algo training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
mae1=mean_absolute_error(y_test,y_pred)
```

```python
mae1
```

```python
mse1=mean_squared_error(y_test,y_pred)
```

```python
mse1
```

```python
#therefor we conclude that the error rates of randonforestregressor is less
#so it is best and efficient way of analys the data with less error rates.
#mae = .44 and mse = .35 when comes to other algo mae1 = .50 mse1 = .41
```
